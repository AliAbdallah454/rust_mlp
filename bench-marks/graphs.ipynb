```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tensor_benchmark.csv")

sizes = df["Size"]
sequential_time = df["Sequential"]
parallel_speedup = df["Parallel Speedup"]
simd_speedup = df["SIMD Speedup"]
parallelsimd_speedup = df["ParallelSIMD Speedup"]

parallel_time = sequential_time / parallel_speedup
simd_time = sequential_time / simd_speedup
parallelsimd_time = sequential_time / parallelsimd_speedup

plt.figure()
plt.plot(sizes, [1]*len(sizes), label="Sequential (1x)", marker='o')
plt.plot(sizes, parallel_speedup, label="Parallel", marker='o')
plt.plot(sizes, simd_speedup, label="SIMD", marker='o')
plt.plot(sizes, parallelsimd_speedup, label="Parallel+SIMD", marker='o')
plt.xlabel("Matrix Size")
plt.ylabel("Speedup (vs Sequential)")
plt.title("Speedup Comparison")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.plot(sizes, sequential_time, label="Sequential", marker='o')
plt.plot(sizes, parallel_time, label="Parallel", marker='o')
plt.plot(sizes, simd_time, label="SIMD", marker='o')
plt.plot(sizes, parallelsimd_time, label="Parallel+SIMD", marker='o')
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.title("Execution Time Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "mlp_data.csv"  # Update path if needed
df = pd.read_csv(file_path)

df.columns = df.columns.str.strip()
df['mlp'] = df['mlp'].str.strip()

plt.figure(figsize=(10, 6))
plt.plot(df['mlp'], df['seq'], marker='o', label='Sequential (baseline)', linestyle='--')
plt.plot(df['mlp'], df['parallel'], marker='o', label='Parallel')
plt.plot(df['mlp'], df['simd'], marker='o', label='SIMD')

plt.title('Speedup of Methods vs Sequential (Speedup = 1 for Sequential)')
plt.xlabel('Model Size')
plt.ylabel('Speedup')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
