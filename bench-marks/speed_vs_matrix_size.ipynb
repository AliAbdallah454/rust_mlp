```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionary mapping matrix sizes to file paths
files = {
    "32": "parallel_benchmark_32.csv",
    "64": "parallel_benchmark_64.csv",
    "128": "parallel_benchmark_128.csv",
    "256": "parallel_benchmark_256.csv",
    "512": "parallel_benchmark_512.csv",
    "1024": "parallel_benchmark_1024.csv",
}

# Iterate over each file and generate a plot
for size, path in files.items():
    df = pd.read_csv(path)
    
    plt.figure()
    for mode in df["Mode"].unique():
        subset = df[df["Mode"] == mode]
        plt.plot(subset["Threads"], subset["Time(s)"], marker='o', label=mode)
    
    plt.title(f"Execution Time vs Threads (Matrix Size: {size}x{size})")
    plt.xlabel("Threads")
    plt.ylabel("Time (s)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

plt.show()
```
